```python
import numpy as np
from pydub import AudioSegment
from pydub.utils import ratio_to_db
```

### Loading from file

```python
# from WAV
audio_1 = AudioSegment.from_wav(
    '../split_data/train/id10001/00001.wav'
)
audio_1
```

### Useful

### Joining two audio

```python
audio_1 + audio_1
```

### Scale loudness

```python
audio_louder = audio_1.apply_gain(ratio_to_db(2))
audio_louder
```

```python
audio_louder.max / audio_1.max
```

```python
audio_1.apply_gain(ratio_to_db(0.3))
```

### Mix two audios

```python
audio_2 = AudioSegment.from_wav('../split_data/train/id10002/00001.wav')
audio_2
```

```python
# cut audio_2 to the length of audio_1
# (pydub does things in miliseconds)
audio_2 = audio_2[:len(audio_1)]
audio_2
```

```python
audio_1.overlay(audio_2)
```
